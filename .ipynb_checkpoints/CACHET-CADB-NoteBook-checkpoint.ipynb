# View Contextul Data

### Single day context data visualisation

```python
path_of_single_day_context_file="/Users/deku/Desktop/CACHET-AFDB/FINAL/CACHET-CADB/annotations/P1/a2b3c4@cachet.dk1/1/context.xlsx"
#path_annotation="/Users/deku/Desktop/CACHET-AFDB/FINAL/CACHET-CADB/annotations/P1/a2b3c4@cachet.dk1/1/annotation.csv"
df=pd.read_excel(path_of_single_day_context_file,usecols=[0,1,2, 3, 4, 5, 9,22,23,24,25,26,28,31])
df.index= pd.to_datetime(df["Date abs [yyyy-mm-dd]"])

## list dataframe containing 24 hours context data
df
```

### Plot 24 hours actaivity, BodyPosition, ActivityClass, NonWearTime

```python
df["MovementAcceleration [g](scaled up by 20)"]=df["MovementAcceleration [g]"]*20
df["BodyPosition [0=unknown, 1=lying supine, 2=lying left, 3=lying prone, 4=lying right, 5=upright, 6=sitting/lying, 7=standing]"]=df["BodyPosition []"]
df["ActivityClass [0=unknown, 1=lying, 2=sitting/standing, 3=cycling, 4=slope up, 5=jogging, 6=slope down, 7=walking, 8=sitting/lying, 9=standing]"]=df["ActivityClass []"]
df["NonWearTime [0= wear, 1 non-wear]"]=df["NonWearTime []"]

df.plot(y=["MovementAcceleration [g](scaled up by 20)",], figsize=(40,3),)
plt.legend(loc='upper left', frameon=False, fontsize=25)
df.plot(y=["BodyPosition [0=unknown, 1=lying supine, 2=lying left, 3=lying prone, 4=lying right, 5=upright, 6=sitting/lying, 7=standing]"], figsize=(40,3), color="green")
plt.legend(loc='upper left', frameon=False, fontsize=25)
df.plot(y=["ActivityClass [0=unknown, 1=lying, 2=sitting/standing, 3=cycling, 4=slope up, 5=jogging, 6=slope down, 7=walking, 8=sitting/lying, 9=standing]"], figsize=(40,3),color="black")
plt.legend(loc='upper left', frameon=False, fontsize=25)
df.plot(y=["NonWearTime [0= wear, 1 non-wear]"], figsize=(40,5),)
plt.legend(loc='upper left', frameon=False, fontsize=25)
```
